```python
# Calcul de la disponibilité totale
dispo_int_totale = dispo_alimentaire_population['Disponibilité intérieure'].sum()
formated_dispo_int_totale = "{:,.0f}".format(dispo_int_totale)
print("disponibilité totale en kg =", formated_dispo_int_totale)

#MONTRER A MOHAMED
```

```python

#création d'une boucle for pour afficher les différentes valeurs en fonction des colonnes aliments pour animaux, pertes, nourritures,semence, traitement, autre utilisation

colonnes_a_afficher = ['Aliments pour animaux','Nourriture', 'Pertes',
             'Semences', 'Traitement', 'Autres Utilisations']

# Boucle pour afficher les valeurs en fonction des colonnes spécifiées
for colonne in colonnes_a_afficher:
    valeurs = dispo_alimentaire_population[colonne].sum()
    proportion = ((valeurs / dispo_int_totale) * 100).round(2)
    print(f"Proportion de {colonne}: {proportion}%")
    
```

```python
#graphique 

import matplotlib.pyplot as plt

# Vos données
colonnes_a_afficher = ['Aliments pour animaux', 'Nourriture', 'Pertes',
                      'Semences', 'Traitement', 'Autres Utilisations']

valeurs = []
noms_colonnes = []

# Calcul des proportions et stockage des valeurs et noms de colonnes
for colonne in colonnes_a_afficher:
    proportion = ((dispo_alimentaire_population[colonne].sum() / dispo_int_totale) * 100).round(2)
    valeurs.append(proportion)
    noms_colonnes.append(colonne)

# Création du graphique en secteurs (pie chart)
plt.figure(figsize=(8, 8))
plt.pie(valeurs, labels=noms_colonnes, autopct='%1.1f%%', startangle=140)
plt.title("Répartition de la disponibilité intérieure")
plt.show()
```

```python
#Création d'une liste avec toutes les variables
liste = dispo_alimentaire_population['Produit'].unique()
print(liste)

Cereales = ['Blé' ,'Céréales, Autres' ,'Maïs' ,'Millet' ,'Orge' ,'Riz (Eq Blanchi)' ,'Avoine' ,'Seigle','Sorgho']
```

```python
#Création d'un dataframe avec les informations uniquement pour ces céréales

data_frame_cereales = dispo_alimentaire_population[dispo_alimentaire_population['Produit'].str.strip().isin(Cereales)]
data_frame_cereales.head()

```

```python
import matplotlib.pyplot as plt

# Calculer la somme de la 'Disponibilité intérieure des céréales'
disponibilite_interieure_total = data_frame_cereales['Disponibilité intérieure'].sum()

# Calculer les sommes des différentes catégories
categories = ['Aliments pour animaux', 'Autres Utilisations', 'Nourriture', 'Pertes', 'Semences']
sommes_categories = [data_frame_cereales[cat].sum() for cat in categories]


# Calculer les proportions par rapport à la Disponibilité intérieure
proportions = [(somme_cat / disponibilite_interieure_total) * 100 for somme_cat in sommes_categories]

# Créer un graphique en camembert (pie chart) pour représenter les proportions
labels = categories
sizes = proportions
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ffb3e6']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Répartition mondiale de la disponibilité intérieure des céréales en 2017 ')
plt.show()



```

```python
# Calcul de la proportion de personnes en sous-nutrition par rapport à la population en 2017
personne_sous_nutrition_2017['proportion_pays'] = (personne_sous_nutrition_2017['sous_nutrition'] / personne_sous_nutrition_2017['Population']) * 100
personne_sous_nutrition_2017.head()

```

```python
# Affichage après tri des 10 pays présentant le plus gros taux de personnes sous-nourries
classement_10_premiers_pays_sous_nutrition = personne_sous_nutrition_2017.sort_values('proportion_pays', ascending=False).head(10)

# Création du graphique en barres horizontal
plt.figure(figsize=(14, 8))
bars = plt.barh(y=classement_10_premiers_pays_sous_nutrition['Zone'], width=classement_10_premiers_pays_sous_nutrition['proportion_pays'], color='skyblue')

# Ajouter des étiquettes de données sur les barres
for bar in bars:
    xval = bar.get_width()
    plt.text(xval, bar.get_y() + bar.get_height()/2, f'{xval:.2f}%', va='center', ha='left', fontsize=10, color='navy')

plt.title("Top 10 des Pays avec la plus grande proportion de personnes sous-nourries")
plt.xlabel('% de la population en sous-nutrition')

# Définir l'échelle de l'axe des ordonnées jusqu'à 100
plt.xlim(0, 100)

# Rotation des étiquettes d'axe des X à 45 degrés pour une meilleure lisibilité
plt.xticks(rotation=45, ha='right')

# Afficher le graphique
plt.grid(axis='x')
plt.tight_layout()
plt.show()




```

```python
import textwrap

# Création du graphique
plt.figure(figsize=(12, 6))
bars = plt.bar(height=classement_10_premiers_pays_sous_nutrition['proportion_pays'],
               x=classement_10_premiers_pays_sous_nutrition['Zone'], color='skyblue')

# Afficher les étiquettes des pourcentages (arrondis) au-dessus des barres avec des libellés enveloppés
for bar in bars:
    yval = bar.get_height()
    label = classement_10_premiers_pays_sous_nutrition.loc[classement_10_premiers_pays_sous_nutrition['proportion_pays'] == yval, 'Zone'].values[0]
    wrapped_label = "\n".join(textwrap.wrap(label, width=12))  # Enveloppe les libellés à la largeur de 10 caractères
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.2, f'{round(yval)}%', ha='center', color='navy', fontsize=10)
    plt.text(bar.get_x() + bar.get_width()/2, -2.5, wrapped_label, ha='center', va='center', fontsize=8)

plt.title("Top 10 des Pays avec la plus grande proportion de personnes sous-nourries")
plt.ylabel('% de la population en sous nutrition')

# Afficher le graphique
plt.xticks([])  # Masquer les étiquettes de l'axe des x
plt.grid(axis='y')
plt.tight_layout()  # Ajuste automatiquement l'espacement entre les sous-plots pour éviter les chevauchements
plt.show()




```
